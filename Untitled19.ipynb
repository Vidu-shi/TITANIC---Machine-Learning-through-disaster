## Missing Values

```python
df.isnull().sum()
```

```python
df['Embarked'].mode
```

```python
df['Embarked'].fillna(value = 'S', inplace = True)
```

```python
dataf = df.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis = 'columns')
```

```python
dataf.isnull().sum()
```

## Dummy Variables

```python
df1 = pd.get_dummies(dataf, columns = ['Sex'])
```

```python
df2 = pd.get_dummies(df1, columns = ['Pclass'])
```

```python
df3 = pd.get_dummies(df2, columns = ['Embarked'])
```

```python
df3.head()
```

```python
df4 = df3.drop(['Sex_male','Pclass_1','Embarked_C'], axis = 'columns')
```

## Logistics Regression

```python
#Attributes and Labels 
x_train = df4.iloc[:,1:10]
```

```python
print(x_train)
```

```python
y_train = df4.iloc[:,0]
```

```python
print(y_train)
```

```python
logistic_model = LogisticRegression(max_iter = 200)
```

```python
model = logistic_model.fit(x_train, y_train)
```

```python
model.intercept_
```

```python
model.coef_
```
